```python
def feature_engineering(df):
    df['half_time_result_sum'] = df['HTHG'] + df['HTAG']
    df['final_result_sum'] = df['FTHG'] + df['FTAG']
    #nan'lar sıfır olacak
    df['home_team_prob'] = (df['HST'] / df['HS'])*100
    df['away_team_prob'] = (df['AST'] / df['AS'])*100
    
    return df


df = feature_engineering(df)
    
    
def sifir_5_Alt_Yari(string):
    if string > 0:
        return 0 #
    else:
        return 1 #

def bir_5_Alt_Yari(string):
    if string > 1:
        return 0 #
    else:
        return 1 #
    
def iki_5_Alt_Yari(string):
    if string > 2:
        return 0 #
    else:
        return 1 #
    
df['0.5_under_half'] = df['final_result_sum'].apply(sifir_5_Alt_Yari)
df['1.5_under_half'] = df['final_result_sum'].apply(bir_5_Alt_Yari)
df['2.5_under_half'] = df['final_result_sum'].apply(iki_5_Alt_Yari)

def sifir_5_Alt_Final(string):
    if string > 0:
        return 0 #bitmemıs
    else:
        return 1 #bitmis

def bir_5_Alt_Final(string):
    if string > 1:
        return 0 #bitmemıs
    else:
        return 1 #bitmis
    
def iki_5_Alt_Final(string):
    if string > 2:
        return 0 #bitmemıs
    else:
        return 1 #bitmis

def uc_5_Alt_Final(string):
    if string > 3:
        return 0 #bitmemıs
    else:
        return 1 #bitmis

def dort_5_Alt_Final(string):
    if string > 4:
        return 0 #bitmemıs
    else:
        return 1 #bitmis
    
df['0.5_under_final'] = df['final_result_sum'].apply(sifir_5_Alt_Final)
df['1.5_under_final'] = df['final_result_sum'].apply(bir_5_Alt_Final)
df['2.5_under_final'] = df['final_result_sum'].apply(iki_5_Alt_Final)
df['3.5_under_final'] = df['final_result_sum'].apply(uc_5_Alt_Final)
df['4.5_under_final'] = df['final_result_sum'].apply(dort_5_Alt_Final)

def sifir_5_Ust_Yari(string):
    if string > 0:
        return 1 #
    else:
        return 0 #

def bir_5_Ust_Yari(string):
    if string > 1:
        return 1 #
    else:
        return 0 #
    
def iki_5_Ust_Yari(string):
    if string > 2:
        return 1 #
    else:
        return 0 #

    
df['0.5_above_half'] = df['final_result_sum'].apply(sifir_5_Ust_Yari)
df['1.5_above_half'] = df['final_result_sum'].apply(bir_5_Ust_Yari)
df['2.5_above_half'] = df['final_result_sum'].apply(iki_5_Ust_Yari)

def sifir_5_Ust_Final(string):
    if string > 0:
        return 1 #
    else:
        return 0 #

def bir_5_Ust_Final(string):
    if string > 1:
        return 1 #
    else:
        return 0 #
    
def iki_5_Ust_Final(string):
    if string > 2:
        return 1 #
    else:
        return 0 #

def uc_5_Ust_Final(string):
    if string > 3:
        return 1 #
    else:
        return 0 #

def dort_5_Ust_Final(string):
    if string > 4:
        return 1 #
    else:
        return 0 #
    
df['0.5_above_final'] = df['final_result_sum'].apply(sifir_5_Ust_Final)
df['1.5_above_final'] = df['final_result_sum'].apply(bir_5_Ust_Final)
df['2.5_above_final'] = df['final_result_sum'].apply(iki_5_Ust_Final)
df['3.5_above_final'] = df['final_result_sum'].apply(uc_5_Ust_Final)
df['4.5_above_final'] = df['final_result_sum'].apply(dort_5_Ust_Final)


s1 = pd.to_numeric(df['FTHG'], errors='coerce')
s2 = pd.to_numeric(df['FTAG'], errors='coerce')

m1 = (s1 >0) & (s2 >0)
m2 = (s1 <0) & (s2 <0)

masks=[m1,m2]
vals = [1,0]

df['mutual_goal'] = np.select(masks,vals)

# Change to categoric column to numeric
df.loc[df['FTR']=='H','FTR']=1
df.loc[df['FTR']=='D','FTR']=0
df.loc[df['FTR']=='A','FTR']=2

# Change to categoric column to numeric
df.loc[df['HTR']=='H','HTR']=1
df.loc[df['HTR']=='D','HTR']=0
df.loc[df['HTR']=='A','HTR']=2

#Maç Sonucu ve Alt / Üst (1,5)

def selector_1(row):
    if row['FTR'] == 1 and row['final_result_sum'] < 2 :
        return 1
    else:
        return 0
    
def selector_2(row):
    if row['FTR'] == 2 and row['final_result_sum'] < 2 :
        return 1
    else:
        return 0
    
def selector_3(row):
    if row['FTR'] == 0 and row['final_result_sum'] < 2 :
        return 1
    else:
        return 0
    
def selector_4(row):
    if row['FTR'] == 1 and row['final_result_sum'] > 2 :
        return 1
    else:
        return 0
    
def selector_5(row):
    if row['FTR'] == 2 and row['final_result_sum'] > 2 :
        return 1
    else:
        return 0
    
def selector_6(row):
    if row['FTR'] == 0 and row['final_result_sum'] > 2 :
        return 1
    else:
        return 0
    
df['MS_1_under_1_5'] = df.apply(lambda row : selector_1(row), axis=1)
df['MS_2_under_1_5'] = df.apply(lambda row : selector_2(row), axis=1)
df['MS_0_under_1_5'] = df.apply(lambda row : selector_3(row), axis=1)

df['MS_1_above_1_5'] = df.apply(lambda row : selector_4(row), axis=1)
df['MS_2_above_1_5'] = df.apply(lambda row : selector_5(row), axis=1)
df['MS_0_above_1_5'] = df.apply(lambda row : selector_6(row), axis=1)

#Maç Sonucu ve Alt / Üst (2,5)

def selector_7(row):
    if row['FTR'] == 1 and row['final_result_sum'] < 3 :
        return 1
    else:
        return 0
    
def selector_8(row):
    if row['FTR'] == 2 and row['final_result_sum'] < 3 :
        return 1
    else:
        return 0
    
def selector_9(row):
    if row['FTR'] == 0 and row['final_result_sum'] < 3 :
        return 1
    else:
        return 0
    
def selector_10(row):
    if row['FTR'] == 1 and row['final_result_sum'] > 3 :
        return 1
    else:
        return 0
    
def selector_11(row):
    if row['FTR'] == 2 and row['final_result_sum'] > 3 :
        return 1
    else:
        return 0
    
def selector_12(row):
    if row['FTR'] == 0 and row['final_result_sum'] > 3 :
        return 1
    else:
        return 0
    
df['MS_1_under_2_5'] = df.apply(lambda row : selector_7(row), axis=1)
df['MS_2_under_2_5'] = df.apply(lambda row : selector_8(row), axis=1)
df['MS_0_under_2_5'] = df.apply(lambda row : selector_9(row), axis=1)

df['MS_1_above_2_5'] = df.apply(lambda row : selector_10(row), axis=1)
df['MS_2_above_2_5'] = df.apply(lambda row : selector_11(row), axis=1)
df['MS_0_above_2_5'] = df.apply(lambda row : selector_12(row), axis=1)

#Maç Sonucu ve Alt / Üst (3,5)

def selector_13(row):
    if row['FTR'] == 1 and row['final_result_sum'] < 4 :
        return 1
    else:
        return 0
    
def selector_14(row):
    if row['FTR'] == 2 and row['final_result_sum'] < 4 :
        return 1
    else:
        return 0
    
def selector_15(row):
    if row['FTR'] == 0 and row['final_result_sum'] < 4 :
        return 1
    else:
        return 0
    
def selector_16(row):
    if row['FTR'] == 1 and row['final_result_sum'] > 4 :
        return 1
    else:
        return 0
    
def selector_17(row):
    if row['FTR'] == 2 and row['final_result_sum'] > 4 :
        return 1
    else:
        return 0
    
def selector_18(row):
    if row['FTR'] == 0 and row['final_result_sum'] > 4 :
        return 1
    else:
        return 0
    
df['MS_1_under_3_5'] = df.apply(lambda row : selector_13(row), axis=1)
df['MS_2_under_3_5'] = df.apply(lambda row : selector_14(row), axis=1)
df['MS_0_under_3_5'] = df.apply(lambda row : selector_15(row), axis=1)

df['MS_1_above_3_5'] = df.apply(lambda row : selector_16(row), axis=1)
df['MS_2_above_3_5'] = df.apply(lambda row : selector_17(row), axis=1)
df['MS_0_above_3_5'] = df.apply(lambda row : selector_18(row), axis=1)

#Maç Sonucu ve Alt / Üst (4,5)
def selector_19(row):
    if row['FTR'] == 1 and row['final_result_sum'] < 5 :
        return 1
    else:
        return 0
    
def selector_20(row):
    if row['FTR'] == 2 and row['final_result_sum'] < 5 :
        return 1
    else:
        return 0
    
def selector_21(row):
    if row['FTR'] == 0 and row['final_result_sum'] < 5 :
        return 1
    else:
        return 0
    
def selector_22(row):
    if row['FTR'] == 1 and row['final_result_sum'] > 5 :
        return 1
    else:
        return 0
    
def selector_23(row):
    if row['FTR'] == 2 and row['final_result_sum'] > 5 :
        return 1
    else:
        return 0
    
def selector_24(row):
    if row['FTR'] == 0 and row['final_result_sum'] > 5 :
        return 1
    else:
        return 0
    
df['MS_1_under_4_5'] = df.apply(lambda row : selector_19(row), axis=1)
df['MS_2_under_4_5'] = df.apply(lambda row : selector_20(row), axis=1)
df['MS_0_under_4_5'] = df.apply(lambda row : selector_21(row), axis=1)

df['MS_1_above_4_5'] = df.apply(lambda row : selector_22(row), axis=1)
df['MS_2_above_4_5'] = df.apply(lambda row : selector_23(row), axis=1)
df['MS_0_above_4_5'] = df.apply(lambda row : selector_24(row), axis=1)

def home_goal(row,team):
    if row['HomeTeam'] == team and row['FTHG'] > 0 :
        return row['FTHG']
    else:
        return 0
    
def away_goal(row,team):
    if row['AwayTeam'] == team and row['FTAG'] > 0 :
        return row['FTAG']
    else:
        return 0
############################################################
def home_goal_yedigi(row,team):
    if row['HomeTeam'] == team and row['FTHG'] > 0 :
        return row['FTAG']
    else:
        return 0

def away_goal_yedigi(row,team):
    if row['AwayTeam'] == team and row['FTAG'] > 0 :
        return row['FTHG']
    else:
        return 0
############################################################
def home_win_final(row,team):
    if row['FTR'] == 1 and row['HomeTeam'] == team:
        return 1
    else:
        return 0
    
def home_lose_final(row,team):
    if row['FTR'] == 2 and row['HomeTeam'] == team:
        return 1
    else:
        return 0
    
def home_draw_final(row,team):
    if row['FTR'] == 0 and row['HomeTeam'] == team:
        return 1
    else:
        return 0   

def away_draw_final(row,team):
    if row['FTR'] == 0 and row['AwayTeam'] == team:
        return 1
    else:
        return 0   

def away_win_final(row,team):
    if row['FTR'] == 2 and row['AwayTeam'] == team:
        return 1
    else:
        return 0
    
def away_lose_final(row,team):
    if row['FTR'] == 1 and row['AwayTeam'] == team:
        return 1
    else:
        return 0

def home_win_half(row,team):
    if row['HTR'] == 1 and row['HomeTeam'] == team:
        return 1
    else:
        return 0
    
def home_lose_half(row,team):
    if row['HTR'] == 2 and row['HomeTeam'] == team:
        return 1
    else:
        return 0
    
def home_draw_half(row,team):
    if row['HTR'] == 0 and row['HomeTeam'] == team:
        return 1
    else:
        return 0   

def away_draw_half(row,team):
    if row['HTR'] == 0 and row['AwayTeam'] == team:
        return 1
    else:
        return 0   

def away_win_half(row,team):
    if row['HTR'] == 2 and row['AwayTeam'] == team:
        return 1
    else:
        return 0
    
def away_lose_half(row,team):
    if row['HTR'] == 1 and row['AwayTeam'] == team:
        return 1
    else:
        return 0

def probability(df,team):
    
    df = df[(df['HomeTeam'] == team) | (df['AwayTeam'] == team)]
    oynadigi_mac = df.shape[0]
    #
    df['home_goal'] = df.apply(lambda row : home_goal(row,team), axis=1)
    df['away_goal'] = df.apply(lambda row : away_goal(row,team), axis=1)
    df['sum_goal'] = df['home_goal'] + df['away_goal'] 
    
    ################################################################################
    
    df['home_goal_yedigi'] = df.apply(lambda row : home_goal_yedigi(row,team), axis=1)
    df['away_goal_yedigi'] = df.apply(lambda row : away_goal_yedigi(row,team), axis=1)
    df['sum_yedigi'] = df['home_goal_yedigi'] + df['away_goal_yedigi']
    ################################################################################
    
    
    display(df.head())
            
    #
    df['home_win_final'] = df.apply(lambda row : home_win_final(row,team), axis=1)
    df['home_lose_final'] = df.apply(lambda row : home_lose_final(row,team), axis=1)
    df['home_draw_final'] = df.apply(lambda row : home_draw_final(row,team), axis=1)
    df['away_draw_final'] = df.apply(lambda row : away_draw_final(row,team), axis=1)
    df['away_win_final'] = df.apply(lambda row : away_win_final(row,team), axis=1)
    df['away_lose_final'] = df.apply(lambda row : away_lose_final(row,team), axis=1)
    #
    df['home_win_half'] = df.apply(lambda row : home_win_final(row,team), axis=1)
    df['home_lose_half'] = df.apply(lambda row : home_lose_final(row,team), axis=1)
    df['home_draw_half'] = df.apply(lambda row : home_draw_final(row,team), axis=1)
    df['away_draw_half'] = df.apply(lambda row : away_draw_final(row,team), axis=1)
    df['away_win_half'] = df.apply(lambda row : away_win_final(row,team), axis=1)
    df['away_lose_half'] = df.apply(lambda row : away_lose_final(row,team), axis=1)
    #
    df = df.iloc[:,27:].reset_index(drop=True)
    df = df.sum(axis=0).to_frame().T
    df.insert(0, 'Team', team)
    df.insert(1, 'Oynadigi_Mac', oynadigi_mac)
    
    home = df['home_goal']
    df.drop(labels=['home_goal'], axis=1,inplace = True)
    df.insert(2, 'home_goal', home)
    
    away = df['away_goal']
    df.drop(labels=['away_goal'], axis=1,inplace = True)
    df.insert(3, 'away_goal', away)
    
    home_yedigi = df['home_goal_yedigi']
    df.drop(labels=['home_goal_yedigi'], axis=1,inplace = True)
    df.insert(4, 'home_goal_yedigi', home_yedigi)
    
    away_yedigi = df['away_goal_yedigi']
    df.drop(labels=['away_goal_yedigi'], axis=1,inplace = True)
    df.insert(5, 'away_goal_yedigi', away_yedigi)
    
    ############################################################################
    
    sum = df['sum_goal']
    df.drop(labels=['sum_goal'], axis=1,inplace = True)
    df.insert(6, 'attigi_goal_ortalama', sum)
    
    sum_yedigi = df['sum_yedigi']
    df.drop(labels=['sum_yedigi'], axis=1,inplace = True)
    df.insert(7, 'yedigi_goal_ortalama', sum_yedigi)
    
    cols_to_keep = df.columns[6:].values.tolist()
    appended_data = []
   
    for i in df.columns[6:8]:
        appended_data.append((df[i].loc[0]/oynadigi_mac))
    for i in df.columns[8:]:
        appended_data.append((df[i].loc[0]/oynadigi_mac)*100)    
         
    df = pd.DataFrame(appended_data).T
    temp = df.shape[1]
    #print(temp)

    df.columns = cols_to_keep
    df.insert(0, 'team', team)
    df.insert(1, 'match_played', oynadigi_mac)
    df.insert(2, 'home_goal', home)
    df.insert(3, 'away_goal', away)
    df.insert(4, 'sum_goal', sum)
    df.insert(5, 'concede_goal_home', home_yedigi) #evde yedigi gol
    df.insert(6, 'concede_goal_away', away_yedigi) # deplesmanda yedigi
    df.insert(7, 'concede_sum', sum_yedigi) #yedigi toplam
    
    df_temp = df.iloc[:,10:temp].T
    df_temp = df_temp.sort_values(by=0,ascending = False).T
    
    df_temp.insert(0, 'team', team)
    df_temp.insert(1, 'match_played', oynadigi_mac)
    df_temp.insert(2, 'home_goal', home)
    df_temp.insert(3, 'away_goal', away)
    df_temp.insert(4, 'sum_goal', sum)
    df_temp.insert(5, 'concede_goal_home', home_yedigi) #evde yedigi gol
    df_temp.insert(6, 'concede_goal_away', away_yedigi) # deplesmanda yedigi
    df_temp.insert(7, 'concede_sum', sum_yedigi) #yedigi toplam
    
    df_temp = df_temp.T
    df_temp = df_temp.rename(columns={0: 'Information Card'})
    
    
    return df_temp
```

```python
ALL = 'ALL'

def unique_sorted_values_plus_ALL(array):
    unique = array.unique().tolist()
    unique.sort()
    unique.insert(0, ALL)
    return unique

dropdown_team_1 = widgets.Dropdown(options =  unique_sorted_values_plus_ALL(df.HomeTeam))
dropdown_team_2 = widgets.Dropdown(options =  unique_sorted_values_plus_ALL(df.HomeTeam))

output_team_1 = widgets.Output()
output_team_2 = widgets.Output()

#sort = False

def dropdown_team_eventhandler_1(change):
    output_team_1.clear_output()
    with output_team_1:
        if (change.new == ALL):
            display(df)
        else:
            display(probability(df,change.new))

def dropdown_team_eventhandler_2(change):
    output_team_2.clear_output()
    with output_team_2:
        if (change.new == ALL):
            display(df)
        else:
            display(probability(df,change.new))
            
            
dropdown_team_1.observe(dropdown_team_eventhandler_1, names='value')
dropdown_team_2.observe(dropdown_team_eventhandler_2, names='value')
```
